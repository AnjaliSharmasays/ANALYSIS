```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("breed2.csv")
df=df.dropna(subset=["id","date","deadline_date","deadline_length_days","category","sid"])
category_counts=df["category"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(category_counts.index,category_counts.values,color=["blue","red"])
plt.title("represantion_of_category")
plt.xlabel("category")
plt.ylabel("counts")
plt.tight_layout()
plt.savefig("Bargraph.png")
plt.show()



days_counts=df["category"].value_counts()
plt.figure(figsize=(6,4))
plt.pie(days_counts,labels=days_counts.index,autopct='%1.1f%%')
plt.title("percentageofcategory")

plt.tight_layout()
plt.savefig("Piechart.png")
plt.show()




x=df["id"]
y=df["deadline_length_days"]
plt.plot(y,x,color="green",marker="o")
plt.title("id vs deadline_length_days")
plt.xlabel("id")
plt.ylabel("deadline_length_days")
plt.tight_layout()
plt.savefig("image3.png")
plt.show()
```
